```python
# 1. 환경설정
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

nltk.download('stopwords')
nltk.download('punkt')
nltk.download('punkt_tab')

# 1. 데이터 로드
train = pd.read_csv('C:/Users/허재민/Desktop/mercari-price-suggestion-challenge/sample_submission.csv/train.tsv', sep='\t')
test  = pd.read_csv('C:/Users/허재민/Desktop/mercari-price-suggestion-challenge/sample_submission.csv/test.tsv',  sep='\t')

print(train.shape)
print(train.head())
```

### 2. 데이터 전처리 (결측치, 이상치)

```python
# 전체 결측치 현황 한눈에 보기
print("=== 결측치 현황 ===")
print(train.isna().sum())
```

```python
# 세부 컬럼별 결측치 확인
# 'category_name' 컬럼

print(f"\n▣ category_name 결측치 수: {train['category_name'].isna().sum()}")
print(f"▣ 전체 대비 비율: {train['category_name'].isna().mean()*100:.2f}%")

# 결측치 샘플 확인 - 어떤 상품들인지 보기
null_category = train[train['category_name'].isna()][
    ['name', 'brand_name', 'item_description', 'price']
].sample(20, random_state=42)

print("\n=== category_name 결측 샘플 ===")
print(null_category.to_string())
```

```python
# 'brand_name' 컬럼

print(f"\n▣ brand_name 결측치 수: {train['brand_name'].isna().sum()}")
print(f"▣ 전체 대비 비율: {train['brand_name'].isna().mean()*100:.2f}%")

# 결측치 샘플 확인 - 어떤 상품들인지 보기
null_brand = train[train['brand_name'].isna()][
    ['name', 'category_name', 'item_description', 'price']
].sample(20, random_state=42)

print("\n=== brand_name 결측 샘플 ===")
print(null_brand.to_string())
```

```python
# 'item_description' 컬럼

print(f"\n▣ item_description 결측치 수: {train['item_description'].isna().sum()}")
print(f"▣ 전체 대비 비율: {train['item_description'].isna().mean()*100:.2f}%")

# 결측치 샘플 확인 - 어떤 상품들인지 보기
null_description = train[train['item_description'].isna()][
    ['name', 'category_name', 'brand_name', 'price']
].sample(6, random_state=42)

print("\n=== item_description 결측 샘플 ===")
print(null_description.to_string())
```

```python
# 'item_description' 결측치 : 'No description yet'과 동일한 값으로 처리
train['item_description'] = train['item_description'].fillna("No description yet")
```

```python
# 'brand_name' 결측치 처리 :
# 1. 기존 브랜드명이 기재된 컬럼에는 1, 누락된 곳에는 0
# 2. 결측치처리 : 'No Brand'
train["has_brand"] = train["brand_name"].notna().astype(int)
train["brand_name"] = train["brand_name"].fillna("No Brand")
```

```python
# 'category_name' 결측치 처리 : 'No Category/No category/No category' - 기존 카테고리 분류 형식과 동일하게
train["category_name"] = train["category_name"].fillna("No category/No category/No category")
```

```python
# 카테고리 분리 적용
train[['cat_1', 'cat_2', 'cat_3']] = train['category_name'].str.split('/', n=2, expand=True)

print(train['cat_1'].value_counts().head(15))  # 상위 카테고리 확인 : 카테고리 미분류 항목 포함 총 11개 대분류
```

```python
# 원본 결측 위치 저장
no_category = train[train['category_name'] == 'No category/No category/No category'].copy()
no_brand    = train[train['brand_name'] == 'No Brand'].copy()

print(f"카테고리 없는 상품 수: {len(no_category)}")
print(f"브랜드 없는 상품 수: {len(no_brand)}")
```

```python
# 처리 후 전체 결측치 현황 한눈에 보기
print("=== 결측치 처리 후 데이터 결측치 현황 ===")
print(train.isna().sum())
```

```python
def hhi_count(df):
    share = df["brand_name"].value_counts(normalize=True)
    return (share ** 2).sum()

def hhi_revenue(df):
    revenue = df.groupby("brand_name")["price"].sum()
    share = revenue / revenue.sum()
    return (share ** 2).sum()

targets = {
    "Women": ["Athletic Apparel", "Shoes", "Women's Handbags", "Tops & Blouses", "Jewelry"],
    "Beauty": ["Makeup", "Skin Care", "Fragrance", "Tools & Accessories", "Hair Care"]
}

rows = []

for cat1, cat2_list in targets.items():
    for cat2 in cat2_list:
        subset = train[
            (train["cat_1"] == cat1) &
            (train["cat_2"] == cat2)
        ]
        
        subset_ex = subset[subset["brand_name"] != "No Brand"]
        
        rows.append({
            "cat_1": cat1,
            "cat_2": cat2,
            
            # 노브랜드 포함
            "HHI_count_include": hhi_count(subset),
            "HHI_revenue_include": hhi_revenue(subset),
            
            # 노브랜드 제외
            "HHI_count_exclude": hhi_count(subset_ex) if len(subset_ex)>0 else np.nan,
            "HHI_revenue_exclude": hhi_revenue(subset_ex) if len(subset_ex)>0 else np.nan,
            
            # 가격 구조
            "avg_price": subset["price"].mean(),
            "price_std": subset["price"].std(),
            
            # 노브랜드 비중
            "NoBrand_count_ratio": (subset["brand_name"] == "No Brand").mean(),
            "NoBrand_GMV_ratio": subset.loc[subset["brand_name"]=="No Brand","price"].sum() / subset["price"].sum()
        })

comparison_full = pd.DataFrame(rows)
comparison_full
```

| > 0.25              | 브랜드 과점 |

브랜드 프리미엄 여부 
- revenue > count → 고가 세그먼트 브랜드 장악
- revenue ≈ count → 가격 프리미엄 없음

노브랜드 영향
- NoBrand_GMV_ratio
    - 40% → 노브랜드 중심 시장

- 20~40% → 혼합형

- < 20% → 브랜드 중심 경향

GMV 기준으로 선정한 6개 세부 시장에 대해 거래 및 매출 기준 HHI를 No Brand 포함·제외 조건으로 비교 분석한 결과, Women 카테고리는 전반적으로 브랜드 분산 구조를 보이며 일부 고가 세그먼트에서 브랜드 프리미엄이 관찰되었다. 반면 Beauty는 노브랜드 비중이 높은 구조로, 특히 Skin Care는 브랜드 과점 시장이 아닌 가격 경쟁 중심 시장으로 해석된다. 이에 따라 카테고리별 차별화된 전략 접근이 필요하다.



```python
comparison_full["GMV"] = comparison_full.apply(
    lambda r: train[(train["cat_1"] == r["cat_1"]) & (train["cat_2"] == r["cat_2"])]["price"].sum(),
    axis=1
)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# -----------------------------
# 0) 필요한 컬럼 체크
# -----------------------------
required_cols = ["cat_1", "cat_2", "GMV", "HHI_revenue_exclude"]
for c in required_cols:
    if c not in comparison_full.columns:
        raise ValueError(f"comparison_full에 '{c}' 컬럼이 필요합니다.")

# -----------------------------
# 1) 기준선 설정
# -----------------------------
HHI_CUTOFF = 0.15
GMV_CUTOFF = comparison_full["GMV"].median()  # 6개면 median이 가장 안정적

# -----------------------------
# 2) 산점도 + 기준선 + 라벨
# -----------------------------
plt.figure(figsize=(11, 7))

x = comparison_full["HHI_revenue_exclude"].values
y = comparison_full["GMV"].values
labels = (comparison_full["cat_1"] + " / " + comparison_full["cat_2"]).values

plt.scatter(x, y)

for xi, yi, lab in zip(x, y, labels):
    plt.annotate(lab, (xi, yi), textcoords="offset points", xytext=(6, 6), ha="left")

# 기준선
plt.axvline(HHI_CUTOFF, linewidth=1)
plt.axhline(GMV_CUTOFF, linewidth=1)

# -----------------------------
# 3) 사분면 이름(전략 라벨) 넣기
#    좌표는 현재 축 범위 기반으로 자동 배치
# -----------------------------
xmin, xmax = plt.xlim()
ymin, ymax = plt.ylim()

# 사분면 중앙 좌표 계산
x_left = (xmin + HHI_CUTOFF) / 2
x_right = (HHI_CUTOFF + xmax) / 2
y_bottom = (ymin + GMV_CUTOFF) / 2
y_top = (GMV_CUTOFF + ymax) / 2

# 사분면 텍스트 (보고서 톤으로)
plt.text(
    x_left, y_top,
    "Q2: Large + Competitive\n(Platform / Segmentation)",
    ha="center", va="center"
)
plt.text(
    x_right, y_top,
    "Q1: Large + Concentrated\n(Key Brand Acquisition)",
    ha="center", va="center"
)
plt.text(
    x_left, y_bottom,
    "Q3: Small + Competitive\n(Selective / Niche Play)",
    ha="center", va="center"
)
plt.text(
    x_right, y_bottom,
    "Q4: Small + Concentrated\n(Partner-led / Targeted Entry)",
    ha="center", va="center"
)

# -----------------------------
# 4) 제목/축 라벨
# -----------------------------
plt.title("GMV–HHI 2x2 Strategy Matrix (HHI excludes No Brand)")
plt.xlabel("HHI_revenue_exclude (brand concentration)")
plt.ylabel("GMV (sum of price)")
plt.tight_layout()
plt.show()


# -----------------------------
# 2) No Brand GMV Ratio by Segment
# -----------------------------
plt.figure(figsize=(10, 6))

bar_labels = (comparison_full["cat_1"] + " / " + comparison_full["cat_2"]).values
bar_values = comparison_full["NoBrand_GMV_ratio"].values

plt.bar(bar_labels, bar_values)
plt.xticks(rotation=45, ha="right")

plt.title("No Brand GMV Ratio by Segment")
plt.xlabel("Segment")
plt.ylabel("No Brand GMV Ratio")
plt.tight_layout()
plt.show()
```
